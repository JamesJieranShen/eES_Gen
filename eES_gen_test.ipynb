```python
from eES_gen import eES_Gen
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm, trange
```

```python
generator = eES_Gen()
events = []
for i in trange(100000):
    events.append(generator.genEvent(eThreshold=0, nuThreshold=1))
```

```python
plt.style.use('/media/linux_store/hep/plot_style.mplstyle')
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist([event['nuEnergy'] for event in events], bins=100, histtype='step', color='red', label=r'$E_{\nu}$')
plt.xlabel('Energy [MeV]')
plt.ylabel('Counts')

plt.hist([event['eKE'] for event in events], bins=100, histtype='step', color='blue', label=r'$E_{e}$')
plt.xlim(0, 100)
plt.legend()

plt.subplot(1, 2, 2)
plt.hist([np.dot(event['eDir'], event['sn_direction']) for event in events], bins=100, histtype='step')
plt.xlabel(r'$\hat{d}_{e} \cdot \hat{d_\nu}$')
plt.ylabel('Counts')
plt.xlim(0.5, 1)

plt.tight_layout()
```
